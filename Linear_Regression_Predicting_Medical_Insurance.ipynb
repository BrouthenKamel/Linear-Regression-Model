**Initializing the parameters**

```python
weights = np.random.rand(6)*20
bias = 3000
weights, bias
```

```python
def predict(datapoint, w , b):
  return np.dot(datapoint,w) + b
```

```python
def loss_function(datapoint, target, w, b):
  return ( ( predict(datapoint, w, b) - target ) ** 2 )
```

```python
def cost_function(x_train, y_train, w, b, length):
  cost = 0
  for i in range(length):
    cost += loss_function(x_train[i,:], y_train[i], w, b)
  return cost/(2*length)
```

```python
def update_params(x_train, y_train,length, w, b, learn_rate):
  save_w = w
  update_w = np.zeros(6)
  save_b = b
  update_b = 0
  for j in range(len(w)):
    difference_w = 0
    difference_b = 0
    for i in range(len(y_train)):
      difference_w += ( predict(x_train[i,:], save_w, save_b) - y_train[i] ) * x_train[i,j]
      difference_b += ( predict(x_train[i,:], save_w, save_b) - y_train[i] )
    update_w[j] = w[j] - learn_rate*difference_w/length
    #print("new w = {}".format(update_w))
  update_b = b - learn_rate*difference_b/length
  #print("new b = {}".format(update_b))
  return [update_w, update_b]
```

```python
iterations = 50
alpha = 0.1
max = len(y_train)
for i in range(iterations):
  print("params -> w = {} , b = {}".format(weights,bias))
  print("iteration : {0} cost-function = {1}".format(i, cost_function(x_train, y_train, weights, bias, max)))
  print("/////////////////////////////////////////////////////////////////////////")
  new_params = update_params(x_train, y_train, max, weights, bias , alpha )
  weights = new_params[0]
  bias = new_params[1]

```

Testing the model

```python
for i in range(len(y_test)):
  print(" predicted charge = {} , actual charge = {}".format(predict(x_test[i], weights, bias), y_test[i]))
```
